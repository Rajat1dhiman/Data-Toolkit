2. How does broadcasting work in NumPy?

Ans.  In NumPy , broadcasting is a powerful feature that allows arrays with different shapes to be combined in arithmetic operations. Rather than creating a duplicate array to match shapes, NumPy "broadcasts" the smaller array across the larger one, making computations efficient.

Here's how it works:

1. **Shape Alignment:** NumPy compares the shapes of two arrays starting from the trailing(rightmost) dimensions.

2. **Broadcasting the Smaller Array**: When compatible, NumPy virtually expands the smaller array so that it matches the shape of the larger array during oprations. This expansion is effiecient because no actual data duplicate occurs.

```python
import numpy as np
#Array A of shape(3,1)
A=np.array([[1],[2],[3]])
#ARRay B of sahpe(1,4)
B=np.array([10,20,30,40])

#broadcasting applies here
result =A+B

print(result)
```

4. Explain the use of the groupby()method in Pandas.

Ans.  The **groupby()** method in Pandas is a poerful tool used for grouping and aggregating data in a DataFrame or Series. It splits the data into groups based on some criteria, apllies afunction to each group independently and combines the results. This is particularly useful for tasks like summarizing, analyzing or transforming data.

**Key Steps in **groupby()**:**

1. Split: The data is divided into groups based on the specified key.

```python
import pandas as pd
#creating a sample dataframe

data={
    'Department':["HR","IT","HR","IT","Finance","Finance"],
    'Employee':["A","B","C","D","E","F"],
    'Salary':[50000,60000,55000,65000,70000,75000]
}

df=pd.DataFrame(data)

#sing groupby to calculate the average salary of the dept.

grouped_data=df.groupby('Department')['Salary'].mean()
print(grouped_data)
```

17. Explain the difference between apply() and map() in Pandas.

Ans. In Pandas, both **apply() and map()** are used to apply funstions to data,but they differ in their scope and use cases. Here is a breakdown:

1. **Scope:**
* **map():
          Specifically designed for **Series ** objects
          Applies a function element-wise to each item in the series.
          Cannot be directly used with DataFrames.

* **apply()**:
             Works on both **Series and DataFrames**          
             When used with a Series , it functionality similarly to **map(()**
             When used with a DataFrame,it applies the function along a specified axis(rows or columns)


2. **Flexibilty**
* **map():
         Limited to element-wise operations on a Series.
         Can work with Pyton functions,lambdas or mappings(eg.dictionaries)             

```python
import pandas as pd

#example with map

s=pd.Series([1,2,3,4])
s_mapped=s.map(lambda x:x*2)
print(s_mapped)
```

```python
#apply():
#More flexible and powerful
#can apply functions to rows,columns or individual elements

#example with dataframe

df=pd.DataFrame({"A":[1,2,3],"B":[4,5,6]})
df_applied=df.apply(lambda x:x.sum(),axis=1)#apply along rows

print(df_applied)
```

21. Why is NumPy's array slicing faster than Python's list slicing?

Ans. NumPy's array slicing is faster than Python's list slicing primarily because of how NumPy arrays are designed and implemented. Here are the key reasons:

1. **Efficient Memory Layout:**
* NumPy arrays are stored in contiguos blocks of memory ,meaning all elements are packed together without gaps. This allows for better cache performance and reduces overhead during slicing operations.

* Python lists, on the other hand , are arrays of pointer of objects, which can lead to scattered memory access and slower performance.

2. **Homogenous Data:**

* NumPy arrays holds elements of the same data type, enabling operations to be carried out at the machine level. This uniformity simplifies slicing and other computations.

* Python lists can cintain objects of various data types, intoducing additional overhead when accessing elements.

3. **Vectorized Operations**:
*  NumPy operations, including slicing are implemented in C and Fortran , which are highly optimized for numerical computations. These low-level implementations make slicing operations faster.

* Python's list slicing, while functional ,is not optimized for numerical operations at such a low level.

4. **Avoidance of Type Checking**:
* NumPy avoids the repeated type-checking that puthon lists require during slicing because all elements in a NumPy array are of the same type.

```python
import plotly.express as px
import pandas as pd

# Sample data for 3D scatter plot
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [5, 6, 7, 8, 9],
    'Z': [10, 11, 12, 13, 14],
    'Color': ['red', 'green', 'blue', 'orange', 'purple']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Color', title="3D Scatter Plot")

# Show the plot
fig.show()
```
